```python
import pandas
import tensorflow as tf
import pandas as pd
import numpy as np
import math

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
```

```python
inputData = pd.read_csv('California Housing Data.csv', delimiter=',')
```

```python
inputData
```

```python
inputData.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
y = inputData.medianHouseValue
X = inputData.drop(['medianHouseValue'],axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
transformedTRAINData = scaler.transform(X_train)
transformedTESTData = scaler.transform(X_test)
```

```python
feat_cols =[]
for i in X_train.columns:
    feat_cols.append(tf.feature_column.numeric_column(i,shape=[1]))
    
feat_cols
```

```python
estimator = tf.estimator.DNNRegressor(feature_columns=feat_cols,hidden_units=[9,7,8,9,7,6,6,9])
input_func = tf.estimator.inputs.pandas_input_fn(X_train,y_train,batch_size=1000,num_epochs=100,shuffle = False,)
estimator.train(input_fn = input_func,steps=10000)
```

```python
predict_func = tf.estimator.inputs.pandas_input_fn(x = X_test,batch_size=100,num_epochs=1,shuffle = False)
predicts = estimator.predict(input_fn = predict_func)
predicts = [i['predictions'] for i in list(predicts)]
```

```python
RMSE = math.sqrt(mean_squared_error(list(predicts), y_test))
RMSE
```
