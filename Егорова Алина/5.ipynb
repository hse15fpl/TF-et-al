### THE DATA

** Read in the census_data.csv data with pandas**

```python
import pandas as pd
import tensorflow as tf
```

```python
census = pd.read_csv('census_data.csv')
print(census.shape)
print(census.columns)
```

```python
census
```

** TensorFlow won't be able to understand strings as labels, you'll need to use pandas .apply() method to apply a custom function that converts them to 0s and 1s. This might be hard if you aren't very familiar with pandas, so feel free to take a peek at the solutions for this part.**

** Convert the Label column to 0s and 1s instead of strings.**

```python
def convert(x):
    if x == ' <=50K':
        return 0
    else:
        return 1 #>50K
```

```python
census['income_bracket'] = census['income_bracket'].apply(lambda x: convert(x))
```

```python
census.head()
```

### Perform a Train Test Split on the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_data = census.drop('income_bracket',axis=1)
y_labels = census['income_bracket']
x_train, x_test, y_train, y_test = train_test_split(x_data,y_labels,test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
x_test
```

### Create the Feature Columns for tf.esitmator

** Take note of categorical vs continuous values! **

```python
census.columns
```

```python
vocab = {} # column: vocabulary list
for column in ['gender', 'occupation', 'marital_status', 'relationship', 'education', 'workclass', 'native_country']:
    vocab[column] = list(set(census[column].values))
print(vocab)
```

** Create the tf.feature_columns for the categorical values. Use vocabulary lists or just use hash buckets. **

```python
gender = tf.feature_column.categorical_column_with_vocabulary_list("gender", vocab['gender'])
occupation = tf.feature_column.categorical_column_with_vocabulary_list("occupation", vocab["occupation"])
marital_status = tf.feature_column.categorical_column_with_vocabulary_list("marital_status", vocab["marital_status"])
relationship = tf.feature_column.categorical_column_with_vocabulary_list("relationship", vocab["relationship"])
education = tf.feature_column.categorical_column_with_vocabulary_list("education", vocab["education"])
workclass = tf.feature_column.categorical_column_with_vocabulary_list("workclass", vocab["workclass"])
native_country = tf.feature_column.categorical_column_with_vocabulary_list("native_country", vocab["native_country"])
```

** Create the continuous feature_columns for the continuous values using numeric_column **

```python
age = tf.feature_column.numeric_column("age")
education_num = tf.feature_column.numeric_column("education_num")
capital_gain = tf.feature_column.numeric_column("capital_gain")
capital_loss = tf.feature_column.numeric_column("capital_loss")
hours_pw = tf.feature_column.numeric_column("hours_per_week")
```

** Put all these variables into a single list with the variable name feat_cols **

```python
feat_cols = [gender,occupation,marital_status,relationship,education,workclass,native_country,
            age,education_num,capital_gain,capital_loss,hours_pw]
```

### Create Input Function

** Batch_size is up to you. But do make sure to shuffle!**

```python
input_fn = tf.estimator.inputs.pandas_input_fn(x=x_train,y=y_train,batch_size=128,num_epochs=100,shuffle=True)
```

#### Create your model with tf.estimator

**Create a LinearClassifier.(If you want to use a DNNClassifier, keep in mind you'll need to create embedded columns out of the cateogrical feature that use strings, check out the previous lecture on this for more info.)**

```python
model = tf.estimator.LinearClassifier(feature_columns=feat_cols)
```

** Train your model on the data, for at least 5000 steps. **

```python
model.train(input_fn=input_fn,steps=2500)
```

### Evaluation

** Create a prediction input function. Remember to only supprt X_test data and keep shuffle=False. **

```python
pred_fn = tf.estimator.inputs.pandas_input_fn(x=x_test,y=y_test, batch_size=128, num_epochs=1, shuffle=False)
```

```python
model.evaluate(pred_fn)
```

** Use model.predict() and pass in your input function. This will produce a generator of predictions, which you can then transform into a list, with list() **

```python
predictions = list(model.predict(input_fn=pred_fn))
predictions
```

** Create a list of only the class_ids key values from the prediction list of dictionaries, these are the predictions you will use to compare against the real y_test values. **

```python
pred_classes = [pred['class_ids'][0] for pred in predictions]
print(len(pred_classes))
print(pred_classes[:10])
```

** Import classification_report from sklearn.metrics and then see if you can figure out how to use it to easily get a full report of your model's performance on the test data. **

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred_classes))
```
